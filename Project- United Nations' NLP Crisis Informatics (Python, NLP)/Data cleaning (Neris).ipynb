```
neris_alldisasters24 = pd.read_csv("/content/drive/MyDrive/UNDP - Crisis Text Informatics/neris_otherdisaster_feb24.csv", error_bad_lines = False)
```

```
all_disasters = pd.concat([neris_earthquake, maryam_earthquake, neris_other_disasters, neris_alldisasters24], ignore_index = True)
all_disasters.shape
```

```
all_disasters_wout_duplicate = all_disasters.drop_duplicates("Tweet", ignore_index = True)
all_disasters_wout_duplicate.shape
```

```
#Also address the new &gt thing that comes up in new other disasters dataset. 

import re

import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stop_words = stopwords.words("english")
#updated the stopwords by adding a list of terms, like km and mi, also the directions from 16 wind compass rose https://en.wikipedia.org/wiki/Points_of_the_compass
newStopWords = ['km','mi','utc','gmt','n','ne','nw','se','sw','s','w','e','nne','ene','ese','sse','ssw','wsw','wnw','nnw']
stop_words.extend(newStopWords)


emoji_pattern = re.compile("["
         u"\U0001F600-\U0001F64F"  # emoticons
         u"\U0001F300-\U0001F5FF"  # symbols & pictographs
         u"\U0001F680-\U0001F6FF"  # transport & map symbols
         u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
         u"\U00002702-\U000027B0"
         u"\U000024C2-\U0001F251"
         "]+", flags=re.UNICODE)

def clean_text(twt):

    cln_twt = re.sub("^b", "", twt) #Remove the extra "b'" of every tweet at the beginning
    cln_twt = re.sub(r"\\n", " ", cln_twt)
    cln_twt = re.sub("&gt;", " ", cln_twt)
    cln_twt = emoji_pattern.sub(r"", cln_twt) #Removing emoji patterns compiled above
    cln_twt = re.sub('((www\.[^\s]+)|(https?://[^\s]+))', '', cln_twt) #Removing the links
    cln_twt = re.sub("(#[^\s]+\s+)*(#[^\s]+)$", "", cln_twt)  #Remove the hashtags at the end of the tweet
    cln_twt = re.sub(r'#', '', #Replace hashtags in the actual tweet with space
                     cln_twt)
    cln_twt = re.sub("[^\d\s]*[\d]+[^\d\s]*", "", cln_twt) #Removes all sorts of numbers and date time characters 
    cln_twt = re.sub(r'[^\w\d\s]', ' ', cln_twt) #Punctuation removed  
    cln_twt = re.sub(r'\s+', ' ', cln_twt) #Too much space replaced by single space
    cln_twt = re.sub(r'^\s+|\s+?$', '', cln_twt.lower()) #Get rid of spaces at the beginning and at the end.

    return ' '.join(
    term for term in cln_twt.split(" ")
    if term not in set(stop_words)
    )
```

```
all_disasters_wout_duplicate["Cleaned Tweet 1"] = all_disasters_wout_duplicate.apply(lambda a: clean_text(a["Tweet"]), axis = 1)
all_dis_wout_dup_reshuff = all_disasters_wout_duplicate.sample(frac = 1, random_state = 191).reset_index(drop = True)

for i in range(0,20):

  print(all_dis_wout_dup_reshuff["Tweet"][i])
```
