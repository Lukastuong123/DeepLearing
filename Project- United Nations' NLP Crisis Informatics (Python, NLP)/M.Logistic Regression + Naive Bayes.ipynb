```python
import pandas as pd

comb_data = pd.read_csv("/content/MyDrive/MyDrive/UNDP - Crisis Text Informatics/combined_dataset.csv", engine='python', encoding='ISO-8859-1', error_bad_lines=False)
comb_data.shape
```

```python
comb_data.head()
```

```python
y = comb_data["target"]
cleaned_tweet = comb_data["Cleaned Tweet 1"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split  
```

Obtaining TF-IDF Vectorizations

```python
import numpy as np

cleaned_tweet = cleaned_tweet[cleaned_tweet != np.nan]
cleaned_tweet.shape
```

```python
tfidf_tweet = TfidfVectorizer()
tfidf_vec_tweet = tfidf_tweet.fit_transform(cleaned_tweet.values.astype("U"))
```

```python
print(type(tfidf_vec_tweet))
print(tfidf_vec_tweet.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(tfidf_vec_tweet, y, test_size = 0.2)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape) 
```

Training the Models

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred_lr))
```

```python
mnb = MultinomialNB()
mnb.fit(X_train, y_train)

y_pred_mnb = mnb.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_mnb))
```
